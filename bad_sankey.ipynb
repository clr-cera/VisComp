```python
df = pd.read_csv("/content/DataCoSupplyChainDataset.csv", encoding="latin-1")
df.columns
```

```python
# df = df.dropna()
df = df.loc[:50000]
node_list = list(df["Customer City"].unique()) + list(df["Department Name"].unique())

source = []
target = []
value = []

for i, r in df.iterrows():
    source.append(node_list.index(r["Department Name"]))
    target.append(node_list.index(r["Customer City"]))
    value.append(r["Order Item Total"])


# print(node_list)
# print(source)
# print(target)
# print(value)

fig = go.Figure(data=[go.Sankey(
    node = dict(
        pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
        label = node_list
    ),
    link = dict(
      source = source,
      target = target,
      value = value
  )
)])



fig.update_layout(title_text="Fluxo de vendas por setor e cidade", font_size=10)
fig.show()
```

```python
# Set transparent background before exporting
# !pip install -U kaleido
fig.update_layout(
    paper_bgcolor='rgba(255,255,255,0)',  # Transparent background
    plot_bgcolor='rgba(255, 255,255,0)'    # Transparent plot area
)

# Export as PNG with transparent background
fig.write_image("sankey_diagram.png")
```
